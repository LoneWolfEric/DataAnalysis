## Calculate pi

```python
import scipy.integrate as integrate
```

```python
x = np.linspace(-1, 1, 1000)
f = lambda x : (1 - x ** 2) ** 0.5
plt.figure(figsize=(6, 6))
plt.plot(x, f(x), x, -f(x))
```

```python
pi_half, error = integrate.quad(f, -1, 1)
pi = pi_half * 2
pi
```
