## 3D图形

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
x1 = [1, 2, 3, 4]
y1 = [1, 2, 3, 4, 5]

X1, Y1 = np.meshgrid(x1, y1)
display(x1, X1, y1, Y1)
```

```python
x = np.arange(0, 7, 0.01)
y = np.arange(-3.5, 3.5, 0.01)

X, Y = np.meshgrid(x, y)
plt.plot(X, Y)

def km_Z(X, Y):
    return  2 + X + Y ** 2

Z = km_Z(X, Y)
display(X.shape, Y.shape, Z.shape)
```

```python
plt.figure(figsize=(20,9))
axes = plt.subplot(1,2,1, projection = '3d')
s = axes.plot_surface(X, Y, Z, rstride = 1000, cstride = 1000, cmap = 'rainbow')
plt.colorbar(s, shrink = 0.5)
```

## 玫瑰图

```python
def show_rose(values, title):
    n = 8
    angle = np.arange(0, 2 * np.pi, 2 * np.pi / n)
    
    radius = np.array(values)
    plt.axes([0, 0, 1, 1], polar = True)
    
    plt.bar(angle, radius)
```

```python
v = [1,2,3,4,5,6,7,8]
show_rose(v, 'test')
```
